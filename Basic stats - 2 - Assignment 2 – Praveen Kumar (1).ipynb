```python
import numpy as np
import scipy.stats as stats

# Given data
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85,
    1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29
])

n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)   # sample standard deviation

mean, std, n
```

```python
# Confidence level
confidence = 0.99
alpha = 1 - confidence

# Degrees of freedom
df = n - 1

# t critical value
t_critical = stats.t.ppf(1 - alpha/2, df)

t_critical
```

```python
# Margin of Error
margin_error = t_critical * (std / np.sqrt(n))

# Confidence Interval
lower = mean - margin_error
upper = mean + margin_error

lower, upper
```

```python
# Known population standard deviation
pop_std = 0.2

# Z critical value for 99%
z_critical = stats.norm.ppf(1 - alpha/2)

# Margin of error (Z)
margin_error_z = z_critical * (pop_std / np.sqrt(n))

# Confidence Interval (Z)
lower_z = mean - margin_error_z
upper_z = mean + margin_error_z

z_critical, lower_z, upper_z
```

```python
CONCLUSION:

In this assignment, a 99% confidence interval for the mean durability of print-heads was constructed using both t-distribution and Z-distribution. The t-distribution was used when population standard deviation was unknown, while Z-distribution was used when it was known. The results show that knowing population standard deviation reduces uncertainty and produces a narrower confidence interval.
```
