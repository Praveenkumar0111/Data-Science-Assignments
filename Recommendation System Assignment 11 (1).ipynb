```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("anime.csv")
df.head()
```

```python
df = df[['name', 'genre', 'rating']]   # keep important columns
df.dropna(inplace=True)
df.head()
```

```python
df['genre'] = df['genre'].str.replace(',', ' ')
```

```python
cv = CountVectorizer(stop_words='english')
matrix = cv.fit_transform(df['genre'])
```

```python
cosine_sim = cosine_similarity(matrix)
cosine_sim
```

```python
def recommend_anime(anime_name):
    idx = df[df['name'] == anime_name].index[0]
    
    scores = list(enumerate(cosine_sim[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    
    scores = scores[1:6]   # top 5
    
    anime_indices = [i[0] for i in scores]
    
    return df['name'].iloc[anime_indices]
```

```python
recommend_anime("Naruto")
```

Conclusion:
A recommendation system was built using cosine similarity
Anime genres were converted into numerical vectors
Similarity between anime was calculated using cosine similarity
Based on similarity scores, top similar anime were recommended
The system successfully recommended anime similar to Naruto
