```python
# H0 (Null): Weekly cost = model cost
# H1 (Alt): Weekly cost is higher than model
```

```python
# Given values
sample_mean = 3050
mu = 1000 + 5*600   # theoretical mean
sigma = 5*25        # std deviation
n = 25

# Z-test
z = (sample_mean - mu) / (sigma / (n**0.5))

z
```

```python
# Critical value at 5% level (one-tailed)
critical_value = 1.645

critical_value
```

```python
if z > critical_value:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

```python
# Hypothesis
H₀: μ = 4000 (weekly cost follows model)
H₁: μ > 4000 (weekly cost is higher)

# Test Statistic
Z = −38

# Critical Value
At α = 0.05 (one-tailed): Z = 1.645

# Decision
Since:
-38 < 1.645

We Fail to Reject H₀

# Final Conclusion (Write This)

Based on the hypothesis test, we fail to reject the null hypothesis at the 5% significance level. There is no strong statistical evidence to support the restaurant owners' claim that weekly operating costs are higher than the model suggests.
```
